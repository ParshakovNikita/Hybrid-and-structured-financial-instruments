# Оценки для опционов колл на отношение цен

Пример (Haug, 2007, с. 203). Вычислите оценки для опционов колл на отношение для
$\sigma_1=0.3$, $\sigma_2=0.4$, $b_1=0.05$,
$b_2=0.03$, $r=0.07$, $S_1=130$, $S_2=100$, $T=0.25, 0.5$, $\rho=\{-0.5, 0, 0.5\}$ и $K=0.1, 0.2, \ldots, 1.0, 2.0, 3.0$, безрисковая ставка 7\%.

```python
sigma1 = 0.3
sigma2 = 0.4
b1 = 0.05
b2 = 0.03
r = 0.07
S1 = 130
S2 = 100
T_range = [0.25, 0.5]
rho_range = [-0.5, 0, 0.5]
K_range = list(np.arange(0.1, 3.1, 0.1))
```

```python
# оценка стоимости опциона колл на отношение цен двух активов
def quotient_option_price_Zhang(S1, S2, b1, b2, sigma1, sigma2, K, rho, T, is_call):
    theta = 1
    if is_call == False:
        theta = -1
    sigma_hat = np.sqrt(sigma1 ** 2 + sigma2 ** 2 - 2 * rho * sigma1 * sigma2)

    d1 = (np.log(S1 / S2 / K) + (b1 - b2 - 0.5 * (sigma1 ** 2 - sigma2 ** 2)) * T) / \
         (sigma_hat * np.sqrt(T))
    d2 = d1 + sigma_hat * np.sqrt(T)
    F = S1 / S2 * np.exp((b1 - b2 +  sigma2 * (sigma2 - rho * sigma1)) * T)
    option = theta * np.exp(-r * T) * (F * norm.cdf(theta * d2) - K * norm.cdf(theta * d1))

    return option
```

```python
outperformance_df = pd.DataFrame([
    {
        'K': K,
        'rho': rho,
        'T': T,
        'price':quotient_option_price_Zhang(S1, S2, b1, b2, sigma1, sigma2, K, rho, T, is_call = True)
    }
    for K in K_range
    for rho in rho_range
    for T in T_range
])
```

### Put on the Minimum of Two Assets
**Payoff**:  
$ \max[X - \min(S_1, S_2), 0] $

**Price**:  
$ p_{\text{min}}(S_1, S_2, X, T) = X e^{-rT} - c_{\text{min}}(S_1, S_2, 0, T) + c_{\text{min}}(S_1, S_2, X, T), $

где:  
$ c_{\text{min}}(S_1, S_2, 0, T) = S_1 e^{(b_1 - r)T} N(d) + S_2 e^{(b_2 - r)T} N(d - \sigma \sqrt{T}), $

---

### Put on the Maximum of Two Assets
**Payoff**:  
$ \max[X - \max(S_1, S_2), 0] $

**Price**:  
$ p_{\text{max}}(S_1, S_2, X, T) = X e^{-rT} - c_{\text{max}}(S_1, S_2, 0, T) + c_{\text{max}}(S_1, S_2, X, T), $

где:  
$ c_{\text{max}}(S_1, S_2, 0, T) = S_2 e^{(b_2 - r)T} + S_1 e^{(b_1 - r)T} N(d) - S_2 e^{(b_2 - r)T} N(d - \sigma \sqrt{T}). $


```python
import numpy as np
from scipy.stats import norm, multivariate_normal

def calculate_d(S1, S2, b1, b2, sigma1, sigma2, rho, T):
    """Вычисляет вспомогательные параметры d, rho1, rho2 и sigma"""
    sigma = np.sqrt(sigma1**2 + sigma2**2 - 2 * rho * sigma1 * sigma2)
    rho1 = (sigma1 - rho * sigma2) / sigma
    rho2 = (sigma2 - rho * sigma1) / sigma
    d = (np.log(S1 / S2) + (b1 - b2 + sigma**2 / 2) * T) / (sigma * np.sqrt(T))
    return d, rho1, rho2, sigma

def calculate_y(S, X, b, sigma, T):
    """Вычисляет параметры y для расчетов"""
    return (np.log(S / X) + (b + sigma**2 / 2) * T) / (sigma * np.sqrt(T))

def c_max_0(S1, S2, b1, b2, r, sigma, d, T):
    """Вычисляет c_max(S1, S2, 0, T)"""
    term1 = S2 * np.exp((b2 - r) * T) + S1 * np.exp((b1 - r) * T) * norm.cdf(d)
    term2 =  - S2 * np.exp((b2 - r) * T) * norm.cdf(d - sigma * np.sqrt(T))
    return term1 + term2

def bivariate_cdf(y, d, rho):
    """Вычисляет двумерную нормальную функцию распределения"""
    mean = [0, 0]  # Средние значения для стандартного нормального распределения
    cov = [[1, rho], [rho, 1]]  # Ковариационная матрица
    return multivariate_normal.cdf([y, d], mean=mean, cov=cov)


def c_max_X(S1, S2, X, b1, b2, r, sigma, sigma1, sigma2, d, y1, y2, rho, rho1, rho2, T):
    """Вычисляет c_max(S1, S2, X, T)"""
    M = bivariate_cdf  # Используем двумерное нормальное распределение
    term1 = S1 * np.exp((b1 - r) * T) * M(y1, d, rho1)
    term2 = S2 * np.exp((b2 - r) * T) * M(y2, -d + sigma * np.sqrt(T), rho2)
    term3 = X * np.exp(-r * T) * (1 - M(-y1 + sigma1 * np.sqrt(T), -y2 + sigma2 * np.sqrt(T), rho))
    return term1 + term2 - term3

def price_put_max(S1, S2, X, T, r, b1, b2, sigma1, sigma2, rho):
    """Расчет стоимости пут-опциона на максимум двух активов"""
    d, rho1, rho2, sigma = calculate_d(S1, S2, b1, b2, sigma1, sigma2, rho, T)
    y1 = calculate_y(S1, X, b1, sigma1, T)
    y2 = calculate_y(S2, X, b2, sigma2, T)
    c_max_0_val = c_max_0(S1, S2, b1, b2, r, sigma, d, T)
    c_max_X_val = c_max_X(S1, S2, X, b1, b2, r, sigma, sigma1, sigma2, d, y1, y2, rho, rho1, rho2, T)
    return X * np.exp(-r * T) - c_max_0_val + c_max_X_val

# Пример данных
S1 = 100
S2 = 105
X = 98
T = 0.5
r = 0.05
sigma1 = 0.11
sigma2 = 0.16
b1 = r - 0.06  # r - q1
b2 = r - 0.09  # r - q2
rho = 0.63

# Расчет стоимости пут-опциона на максимум двух активов
price_max = price_put_max(S1, S2, X, T, r, b1, b2, sigma1, sigma2, rho)

print(f"Стоимость пут-опциона на максимум двух активов: {price_max:.4f}") # значения совпадают с расчетами автора
```
